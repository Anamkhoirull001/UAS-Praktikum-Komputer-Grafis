```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
# Baca dataset dengan pandas disini saya memakai variabel data
data = pd.read_csv('Uas_Praktikum.csv')
```

saya membuat 3 grafik histogram untuk Purchase_Price, Current_Price, Gain_Loss dikarenakan pada data Gain_Loss memiliki data yang sangat berbeda dengan 2 kolom sebelumnya

```python
purcise_price = data['Purchase_Price']
print(purcise_price)
plt.figure(figsize=(10,5))
plt.hist(purcise_price, bins='auto',stacked=True, alpha=0.7, color='lightblue' )
# plt.xlabel('Current Price', fontsize = 10)
plt.ylabel('Frequency',fontsize = 10)
plt.title('1. a, Histogram of Purchase Price',fontsize=20, color='orange')
plt.xlim(xmin=min(purcise_price),xmax=max(purcise_price))
plt.show()
```

```python
curren_price = data['Current_Price']
gain_loss = data['Gain_Loss']
print(purcise_price)
plt.figure(figsize=(10,5))
plt.hist(curren_price, bins='auto', alpha=0.7, color='lightblue' )
# plt.xlabel('Current Price', fontsize = 10)
plt.ylabel('Frequency',fontsize = 10)
plt.title('1. b, Histogram of Current Price',fontsize=20, color='orange')
plt.xlim(min(curren_price),max(curren_price))
plt.show()
```

```python
curren_price = data['Current_Price']
gain_loss = data['Gain_Loss']
print(gain_loss)
plt.figure(figsize=(10,5))
plt.hist(gain_loss, bins='auto', alpha=0.7, color='lightblue' )
# plt.xlabel('Current Price', fontsize = 10)
plt.ylabel('Frequency',fontsize = 10)
plt.title('1. c, Histogram of Gain Loss',fontsize=20, color='orange')
plt.xlim(min(gain_loss),max(gain_loss))
plt.show()
```

```python
#ambil informasi top 5 berdasarkan Dividend_Share
top5s = (data.groupby('Symbol')['Dividend_Share']
                .sum()
                .reset_index()
                .sort_values(by='Dividend_Share',ascending=False)
                .head(5))
#membuat dataframe baru, hanya top 5
top5 = data[(data['Symbol'].isin(top5s['Symbol'].to_list()))]
```

```python
plt.figure(figsize=(10,5))
plt.hist(top5.groupby('Symbol')['Dividend_Share'].nlargest(5), bins=10, stacked=True, range=(1,3), color='lightblue')
plt.title('2. Histogram of Top 5 Dividend Share',fontsize=20, color='blue')
plt.xlabel('Dividend Share', fontsize = 10)
plt.ylabel('Frequency',fontsize = 10)
plt.xlim(xmin=0.9,xmax=2.5)
```

```python
plt.figure(figsize=(10,5))
# kelompokan berdasar Symbol yang memiliki Market_Value serta sortir top 5 tertinggi dan  urutkan dari terbesar ke terkecil
print(top5.groupby('Symbol')['Market_Value'].nlargest(5).sort_values(ascending=False))
plt.hist(top5.groupby('Symbol')['Market_Value'].nlargest(5), bins=10, stacked=True, range=(7000,15000), color='lightblue')
plt.title('3. Histogram of Top 5 Market Value',fontsize=20, color='blue')
plt.xlabel('Market Value', fontsize = 10)
plt.ylabel('Frequency',fontsize = 10)
plt.xlim(xmin=7000,xmax=15000)
```

```python
#ambil informasi top 7 berdasarkan Annual-Yield
top7s = (data.groupby('Symbol')['Annual_Yield']
                .sum()
                .reset_index()
                .sort_values(by='Annual_Yield',ascending=False)
                .head(7))
#membuat dataframe baru, hanya top 7
top7 = data[(data['Symbol'].isin(top7s['Symbol'].to_list()))]
# kelompokan berdasar Symbol yang memiliki Annual_Yield serta sortir top 7 tertinggi dan  urutkan dari terbesar ke terkecil
print(top7s)
# plt.figure(figsize=(10,7))
plt.hist(top7.groupby('Symbol')['Annual_Yield'].head(7).sort_values(ascending=True), bins=10, stacked=True, range=(0,6), color='lightblue')
plt.title('4. Histogram of Top 7 Annual Yield',fontsize=20, color='blue')
plt.xlabel('Annual Yield', fontsize = 10)
plt.ylabel('Frequency',fontsize = 10)
plt.xlim(xmin=0,xmax=6)
```
